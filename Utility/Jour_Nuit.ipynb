```python
from matplotlib import pyplot as plt
import datetime
import numpy as np
import pandas as pd
import seaborn as sns

data = pd.read_csv('eplusout.csv')
data_i = pd.DataFrame(data)

# Ajout de l'année dans la feature de temps
data_i['Date/Time']=data_i['Date/Time'].apply(lambda x :  x.split(' ')[1]+'/2018 '+ x.split(' ')[3])

# Séparation des données de date et de temps
data_i['Date'] = data_i['Date/Time'].apply(lambda x :  x.split(' ')[0])
data_i['Time'] = data_i['Date/Time'].apply(lambda x :  x.split(' ')[1])
#data_i['Date/Time'] = pd.to_datetime(data_i['Date/Time'],format='%d/%m/%Y %H:%M:%S', errors = 'coerce')
data_i.head()
```

```python
# Création du mask de séparation horaire des Jours/Nuits
mask = (data_i['Time'] > '06:30:00') & (data_i['Time'] <= '18:15:00')
#data_i['Jour/Nuit'] = np.where(mask, 'Jour', 'Nuit')

# Jeu de données avec une nouvelle colonne Jour/Nuit en nombres
data_m = data_i.copy()
data_m['Jour=0/Nuit=1'] = np.where(mask, 0, 1)

# Suppression des colonnes de chaînes de caractères
data_m.pop("Date/Time")
#data_m.pop("Jour/Nuit")
data_m.pop("Date")
data_m.pop("Time")

# Représentation de la moyenne de chaque colonne selon la période de la journée
data_m.groupby('Jour=0/Nuit=1').mean()
```

```python
# Création des jeux de données séparés (Jour/Nuit)
data_nuit = data_m[data_m['Jour=0/Nuit=1']==1]
data_jour = data_m[data_m['Jour=0/Nuit=1']==0]
data_nuit.head()
```

```python
import pandas as pd
import plotly.express as px
data_1 = pd.read_csv('eplusout_1.csv')
data = pd.read_csv('data_missing.csv')
data.head()
```

```python
data_1.head()
```

```python
data_i=pd.DataFrame(data, columns=['Date/Time',
                                'Environment:Site Outdoor Air Drybulb Temperature [C](Hourly)'])
data_i2=pd.DataFrame(data_1, columns=['Date/Time',
                                'Environment:Site Outdoor Air Drybulb Temperature [C](TimeStep)'])
data_i.count()

#data_i.interpolate()
```

```python
data_i2.count()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data_i['Date/Time'], data_i['Environment:Site Outdoor Air Drybulb Temperature [C](Hourly)'], color='red')
plt.title('Outdoor Air Temperature / missing value', fontsize=14)
plt.xlabel('time', fontsize=14)
plt.ylabel('Outdoor Air Temperatur', fontsize=14)
plt.show()
```

```python
fig = px.scatter(data_i, x = 'Date/Time', y = 'Environment:Site Outdoor Air Drybulb Temperature [C](Hourly)', title='Outdoor Temperaturemissing value ')
fig.update_traces(name='Temp/miss_val', showlegend = True)
fig.add_scatter(x=data_i['Date/Time'].interpolate(), y=data_i['Environment:Site Outdoor Air Drybulb Temperature [C](Hourly)'].interpolate(),name='Temp/interpolation', mode='lines')

fig.show()
```

```python
fig = px.line(data_i.interpolate(), x = 'Date/Time', y = 'Environment:Site Outdoor Air Drybulb Temperature [C](Hourly)', title='Outdoor Temperature missing value correction')
fig.update_traces(name='Temp/Rel', showlegend = True)
fig.add_scatter(x=data_i2['Date/Time'], y=data_i2['Environment:Site Outdoor Air Drybulb Temperature [C](TimeStep)'],name='Temp/interpolation', mode='lines')

fig.show()
```

```python

error_inter = abs(data_i2['Environment:Site Outdoor Air Drybulb Temperature [C](TimeStep)']- data_i.interpolate()['Environment:Site Outdoor Air Drybulb Temperature [C](Hourly)']) 
fig2 = px.line(data_i.interpolate(), x = 'Date/Time', y = error_inter, title='Outdoor Temperature missing value correction')
fig2.update_traces(name='Erreur quadratique', showlegend = True)
fig2.show()
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(data_i2['Environment:Site Outdoor Air Drybulb Temperature [C](TimeStep)'], data_i.fillna(0).interpolate()['Environment:Site Outdoor Air Drybulb Temperature [C](Hourly)'] ))
```
